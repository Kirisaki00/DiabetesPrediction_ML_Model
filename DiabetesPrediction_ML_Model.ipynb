```python
df["Pregnancies"]=df["Pregnancies"].replace(0,df["Pregnancies"].median())
df["Glucose"]=df["Glucose"].replace(0,df["Glucose"].mean())
df["BloodPressure"]=df["BloodPressure"].replace(0,df["BloodPressure"].median())
df["SkinThickness"]=df["SkinThickness"].replace(0,df["SkinThickness"].median())
df["Insulin"]=df["Insulin"].replace(0,df["Insulin"].median())
df["BMI"]=df["BMI"].replace(0,df["BMI"].mean())
df["DiabetesPedigreeFunction"]=df["DiabetesPedigreeFunction"].replace(0,df["DiabetesPedigreeFunction"].median())
df["Age"]=df["Age"].replace(0,df["Age"].median())
```

```python
df.head(20)
```

```python
x=df.drop(columns="Outcome",axis=1);
y=df["Outcome"]
```

```python
fig,ax=plt.subplots(figsize=[10,10]);
sns.boxplot(data=x,ax=ax);
plt.show()
```

```python
x.columns
```

```python
cols=['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin','BMI', 'DiabetesPedigreeFunction', 'Age'];
for cols in cols:
    q1=x[cols].quantile(0.25);
    q3=x[cols].quantile(0.75);
    IQR=q3-q1;
    lower_bound=q1-1.5*IQR;
    upper_bound=q3+1.5*IQR;
    mask=(x[cols]>=lower_bound) & (x[cols]<=upper_bound)
```

```python
print(x.shape);
print(y.shape)
```

```python
print(x[mask].shape);
print(y[mask].shape)
x_outlier_removed=x[mask];
y_outlier_removed=y[mask];
```

```python
from sklearn.preprocessing import StandardScaler;
scaler=StandardScaler();
x_scaled=scaler.fit_transform(x[mask])
```

```python
x_scaled.shape
```

```python
x_scaled=pd.DataFrame(x_scaled,columns=['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin','BMI', 'DiabetesPedigreeFunction', 'Age'])
```

```python
fig,ax=plt.subplots(figsize=[15,10]);
sns.boxplot(data=x_scaled,ax=ax);
plt.show()
```

```python
q=x_scaled["Insulin"].quantile(0.93);
x_scaled = x_scaled.reset_index(drop=True)
y_outlier_removed = y_outlier_removed.reset_index(drop=True)
mask=x_scaled["Insulin"]<q
dataNew=x_scaled[mask]
y_outlier_removed=y_outlier_removed[mask]
```

```python
y_outlier_removed.shape
```

```python
fig,ax=plt.subplots(figsize=[15,10]);
sns.boxplot(dataNew,ax=ax);
plt.show()
```

```python
dataNew.shape
```

```python
from sklearn.model_selection import train_test_split;
x_train,x_test,y_train,y_test=train_test_split(dataNew,y_outlier_removed,test_size=.33,random_state=42)
```

```python
y_train.value_counts()
```

```python
from imblearn.over_sampling import SMOTE;
smote=SMOTE(random_state=42);
x_train_resample, y_train_resample = smote.fit_resample(x_train, y_train)
y_train_resample.value_counts()
```

```python
from sklearn.linear_model import LogisticRegression;
classification=LogisticRegression();
classification.fit(x_train_resample,y_train_resample)
```

```python
y_pred=classification.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report;
accuracy_score(y_pred,y_test)
```

```python
target_names=["Non-Diabetic","Diabetic"]
print(classification_report(y_test,y_pred,target_names=target_names))
```
